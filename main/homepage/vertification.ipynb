```python
##라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
%matplotlib inline
```

```python
def make_plt(df,column,units,target):
    fig,ax = plt.subplots(1,1,figsize=(10,5))
    for unit in units:
        sns.kdeplot(x=df[df[column]==unit][target],ax=ax)
    plt.legend(units)
    plt.title(column)
    plt.savefig(os.path.join(".","static","homepage","image","{}_{}.png".format(column.replace("/","_"),target.replace("/","_"))))
    plt.show()


def make_combi(df,target_column,score_columns):
    for column in target_column:
        for score in score_columns:
            make_plt(df,column,np.unique(df[column]),score)
        
```

```python
student_df = pd.read_csv('./static/homepage/data/StudentsPerformance.csv')
make_combi(student_df,['gender','race/ethnicity','parental level of education','lunch','test preparation course'],['math score','reading score','writing score'])
```
